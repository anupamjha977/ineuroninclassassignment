## K-1 one hot encoding

```python
df["Classes"]=pd.get_dummies(df["Classes"],drop_first=True)
```

```python
df.head()
```

```python
x=df.drop(columns="Temperature",axis=1)
y=df["Temperature"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_test.shape
```

# standardization

```python
from sklearn.preprocessing import StandardScaler #standardizing or feature scaling the model
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
x_train.info()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test.info()
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

# Model Creation

### creating the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(x_train,y_train)
```

```python
print(regression.coef_)#printing coefficients
```

```python
print(regression.intercept_)#printing intercepts
```

```python
## PRediction for the test data
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

### Assumption of Linear Regression

### 1.The Two Variables Should be in a Linear Relationship

```python
fig, (ax1, ax2, ax3,ax4, ax5, ax6) = plt.subplots(ncols=6,figsize=(12, 2.5))

ax1.scatter(df['RH'], df['Temperature'])
ax1.set_title("RH")
ax2.scatter(df['Ws'], df['Temperature'])
ax2.set_title("Ws")
ax3.scatter(df['Rain'], df['Temperature'])
ax3.set_title("Rain")
ax4.scatter(df['FFMC'], df['Temperature'])
ax4.set_title("FFMC")
ax5.scatter(df['DMC'], df['Temperature'])
ax5.set_title("DMC")
ax6.scatter(df['ISI'], df['Temperature'])
ax6.set_title("ISI")

plt.show()
```

### 2.Multicollinearity

```python
sns.heatmap(x.corr(),annot=True)#there shouldnt be any multicollinearity between independent feature
```

### 3.Normality of Residual

```python
residual=y_test-reg_pred
```

```python
sns.displot(residual,kind= "kde")#residual should have normal distribution
```

### 4. Homoscedasticity

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

### 5. No Autocorelation of Residual

```python
plt.plot(residual)#residual shouldn't have any autocorelation there plot shouldn't form any pattern
```

```python
## Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# Rsquare and adjusted Rsquare

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,reg_pred)
score
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
